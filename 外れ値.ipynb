```python
import numpy as np
from pandas import DataFrame
```

```python
np.random.seed(12345)
```

```python
df = DataFrame(np.random.randn(1000, 4))
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(10)
```

```python
df.describe()
```

```python
col = df[0]
col.head()
```

```python
# 絶対値が3より大きいデータ
col[np.abs(col)>3]
```

```python
# いずれかの列に絶対値が3以上の行
df[(np.abs(df)>3).any(1)]
```

```python
# データの正負を確認
np.sign(df).head()
```

### やりたいこと
- 絶対値3のデータを外れ値とし、3以上なら3、-3以下なら-3とデータを固定する
